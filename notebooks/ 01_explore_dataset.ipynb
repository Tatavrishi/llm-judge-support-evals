```python
from datasets import load_dataset

# Load the Bitext customer support dataset
# This will download ~20MB the first time, then cache it locally
dataset = load_dataset("bitext/Bitext-customer-support-llm-chatbot-training-dataset")

# Show what we got
print(dataset)
```

```python
# Convert to a pandas DataFrame — easier to explore
import pandas as pd

df = pd.DataFrame(dataset["train"])

# What columns do we have?
print("Columns:", df.columns.tolist())
print("Total rows:", len(df))
print("\nFirst row:")
print(df.iloc[0])
```

```python
# Look at 10 random customer questions and their ideal responses
sample = df.sample(10, random_state=42)

for i, row in sample.iterrows():
    print(f"--- Example {i} ---")
    print(f"INTENT: {row['intent']}")
    print(f"CUSTOMER: {row['instruction']}")
    print(f"IDEAL RESPONSE: {row['response']}")
    print()
```
